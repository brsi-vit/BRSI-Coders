```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler

# Set style for plots
sns.set(style="whitegrid")
```

```python
# Create synthetic data
np.random.seed(42)
data = np.random.normal(50, 10, 100)
data_with_outliers = np.append(data, [150, 160, 170])  # Add some outliers

df = pd.DataFrame({'Original': data_with_outliers})
```

```python
def plot_transformation(original, transformed, title):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.histplot(original, kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title('Before Transformation')

    sns.histplot(transformed, kde=True, ax=axes[1], color='orange')
    axes[1].set_title(f'After {title}')

    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()
```

Min-Max Scaling

```python
scaler = MinMaxScaler()
scaled_minmax = scaler.fit_transform(df[['Original']])

plot_transformation(df['Original'], scaled_minmax, 'Min-Max Scaling')
```

Standardization (Z-score Normalization)

```python
scaler = StandardScaler()
scaled_standard = scaler.fit_transform(df[['Original']])

plot_transformation(df['Original'], scaled_standard, 'Standardization (Z-score)')
```

Robust Scaling

```python
scaler = RobustScaler()
scaled_robust = scaler.fit_transform(df[['Original']])

plot_transformation(df['Original'], scaled_robust, 'Robust Scaling')
```

Max Abs Scaling

```python
scaler = MaxAbsScaler()
scaled_maxabs = scaler.fit_transform(df[['Original']])

plot_transformation(df['Original'], scaled_maxabs, 'Max Abs Scaling')
```
