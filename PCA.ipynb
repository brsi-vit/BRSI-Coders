```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import plotly.express as px
```

```python
df = pd.read_csv(r"C:\Users\panna\Downloads\diabetes.csv")
```

```python
df.shape
```

```python
df.head()
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x = df.drop(columns='Outcome')
y = df['Outcome']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#Scaling or Mean centering
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test) 
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)
```

```python
import pandas as pd
import plotly.express as px

# Merge PCA data with target
pca_df = pd.DataFrame(x_train_pca, columns=['PC1', 'PC2','PC3'])
pca_df['Outcome'] = y_train.reset_index(drop=True)
    
pca_df
```

```python
#Interactive 3D Plot
import plotly.express as px

# Create 3D scatter plot with Plotly
fig = px.scatter_3d(
    pca_df,
    x='PC1',
    y='PC2',
    z='PC3',
    color='Outcome',             # Color by target class
    title="3D Scatter Plot (Interactive)",
    labels={"Outcome": "Class"}
)

# Show the plot
fig.show()
```

```python
# Plot
fig = px.scatter(
    pca_df, x='PC1', y='PC2', color='Target',
    title="Train Data after PCA",
    labels={'Target': 'Class'}
)
fig.show()
```
