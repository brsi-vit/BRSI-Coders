```python
# Required Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# User-Defined Visualization Function for Before & After Transformation (KDE Plot)
def plot_kde_comparison(before, after, title_before="Before Transformation", title_after="After Transformation"):
    plt.figure(figsize=(12, 6))

    # KDE Plot for Before Transformation
    plt.subplot(1, 2, 1)
    sns.kdeplot(before, color='skyblue', shade=True)
    plt.title(f"{title_before} - KDE Plot")
    plt.xlabel('Value')
    plt.ylabel('Density')

    # KDE Plot for After Transformation
    plt.subplot(1, 2, 2)
    sns.kdeplot(after, color='salmon', shade=True)
    plt.title(f"{title_after} - KDE Plot")
    plt.xlabel('Value')
    plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

    # Print Before and After Data Side by Side
    print(f"\n{title_before} Values:")
    print(before)
    print(f"\n{title_after} Values:")
    print(after)
```

1. Polynomial Features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

# Sample Data
data = {'x1': [1, 2, 3, 4, 5], 'x2': [5, 4, 3, 2, 1]}
df = pd.DataFrame(data)

# Polynomial Features (x^2, x^3)
poly = PolynomialFeatures(degree=3)
poly_features = poly.fit_transform(df)

# Polynomial Features DataFrame
poly_features_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(df.columns))

# Function for KDE comparison plot
def plot_kde_comparison(before, after, ax, label_before, label_after):
    sns.kdeplot(before, label=label_before, shade=True, ax=ax)
    sns.kdeplot(after, label=label_after, shade=True, ax=ax)
    ax.legend()
    ax.set_title(f"KDE Plot: {label_before} vs {label_after}")

# Create subplots (side-by-side)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot x vs x^2 and x vs x^3
plot_kde_comparison(df['x1'], poly_features_df['x1^2'], axes[0], "Before Polynomial Transformation (x1)", "After Polynomial Transformation (x1^2)")
plot_kde_comparison(df['x1'], poly_features_df['x1^3'], axes[1], "Before Polynomial Transformation (x1)", "After Polynomial Transformation (x1^3)")

plt.tight_layout()
plt.show()
```

Manually doing polynomial transformation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
data = {'x1': [1, 2, 3, 4, 5], 'x2': [5, 4, 3, 2, 1]}
df = pd.DataFrame(data)

# Manually create Polynomial Features (x^2, x^3)
df['x1^2'] = df['x1'] ** 2
df['x1^3'] = df['x1'] ** 3

# Function for KDE comparison plot
def plot_kde_comparison(before, after, ax, label_before, label_after):
    sns.kdeplot(before, label=label_before, shade=True, ax=ax)
    sns.kdeplot(after, label=label_after, shade=True, ax=ax)
    ax.legend()
    ax.set_title(f"KDE Plot: {label_before} vs {label_after}")

# Create subplots (side-by-side)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot x1 vs x1^2 and x1 vs x1^3
plot_kde_comparison(df['x1'], df['x1^2'], axes[0], "Before Polynomial Transformation (x1)", "After Polynomial Transformation (x1^2)")
plot_kde_comparison(df['x1'], df['x1^3'], axes[1], "Before Polynomial Transformation (x1)", "After Polynomial Transformation (x1^3)")

plt.tight_layout()
plt.show()
```

2. Interaction Terms

```python

# Interaction Term (x1 * x2)
df['interaction'] = df['x1'] * df['x2']

# Visualize KDE Before and After Interaction Term
plot_kde_comparison(df['x1'], df['interaction'], "Before Interaction Term", "After Interaction Term (x1 * x2)")
```

3. Log Transformation

```python
# Log Transformation (Reduces Right Skewness)
df['log_transformed'] = np.log(df['x1'])

# Visualize KDE Before and After Log Transformation
plot_kde_comparison(df['x1'], df['log_transformed'], "Before Log Transformation", "After Log Transformation (log(x1))")
```

4. Square Root and Cube Root Transformation

```python
# Square Root Transformation
df['sqrt_transformed'] = np.sqrt(df['x1'])

# Cube Root Transformation
df['cube_root_transformed'] = np.cbrt(df['x1'])

# Visualize KDE Before and After Square Root Transformation
plot_kde_comparison(df['x1'], df['sqrt_transformed'], "Before Square Root Transformation", "After Square Root Transformation (sqrt(x1))")

# Visualize KDE Before and After Cube Root Transformation
plot_kde_comparison(df['x1'], df['cube_root_transformed'], "Before Cube Root Transformation", "After Cube Root Transformation (cbrt(x1))")
```

5. Exponential Transformation

```python
# Exponential Transformation
df['exp_transformed'] = np.exp(df['x1'])

# Visualize KDE Before and After Exponential Transformation
plot_kde_comparison(df['x1'], df['exp_transformed'], "Before Exponential Transformation", "After Exponential Transformation (exp(x1))")
```

6. Inverse Transformation

```python
# Inverse Transformation
df['inverse_transformed'] = 1 / df['x1']

# Visualize KDE Before and After Inverse Transformation
plot_kde_comparison(df['x1'], df['inverse_transformed'], "Before Inverse Transformation", "After Inverse Transformation (1/x1)")
```

Box-Cox

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import boxcox

# Sample Data
data = {'x1': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Ensure positivity for Box-Cox
shifted_x1 = df['x1'] + 1  # All values already positive, adding 1 just in case
boxcox_transformed, _ = boxcox(shifted_x1)

# Print before and after values side by side
comparison_df = pd.DataFrame({'Original x1': df['x1'], 'Box-Cox x1': boxcox_transformed})
print("Before and After Box-Cox Transformation:")
print(comparison_df)

# Function to plot side-by-side KDE
def plot_kde_comparison(before, after, label_before, label_after):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # KDE of original data
    sns.kdeplot(before, shade=True, ax=axes[0], color='blue')
    axes[0].set_title(label_before)
    axes[0].set_xlabel("Value")

    # KDE of transformed data
    sns.kdeplot(after, shade=True, ax=axes[1], color='green')
    axes[1].set_title(label_after)
    axes[1].set_xlabel("Value")

    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_kde_comparison(df['x1'], boxcox_transformed,
                    "Before Box-Cox Transformation (x1)",
                    "After Box-Cox Transformation (x1)")
```

Yeo-Johnson Transform

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer

# Sample Data (can include negatives and zero)
data = {'x1': [-3, -1, 0, 1, 4, 7, 10]}
df = pd.DataFrame(data)

# Apply Yeo-Johnson Transform
pt = PowerTransformer(method='yeo-johnson')
yeo_johnson_transformed = pt.fit_transform(df[['x1']])

# Create a DataFrame for transformed values
df_transformed = pd.DataFrame(yeo_johnson_transformed, columns=['x1_yeo_johnson'])

# Print original and transformed side by side
comparison_df = pd.concat([df['x1'], df_transformed['x1_yeo_johnson']], axis=1)
print("Before and After Yeo-Johnson Transformation:")
print(comparison_df)

# Function to plot side-by-side KDE
def plot_kde_comparison(before, after, label_before, label_after):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    sns.kdeplot(before, shade=True, ax=axes[0], color='blue')
    axes[0].set_title(label_before)
    axes[0].set_xlabel("Value")

    sns.kdeplot(after, shade=True, ax=axes[1], color='green')
    axes[1].set_title(label_after)
    axes[1].set_xlabel("Value")

    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_kde_comparison(df['x1'], df_transformed['x1_yeo_johnson'],
                    "Before Yeo-Johnson Transformation (x1)",
                    "After Yeo-Johnson Transformation (x1)")
```
